```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
```

```python
url='http://xiaohua.zol.com.cn/lengxiaohua/1.html'
html=requests.get(url,headers=headers)
soup=BeautifulSoup(html.text,'lxml')
datas=soup.select('.article-summary')
for data in datas:
    source=data.select('.article-source')[0].text
    content=data.select('.summary-text')[0].text
    print(source,content)
```
